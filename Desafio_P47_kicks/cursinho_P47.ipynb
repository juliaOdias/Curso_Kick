```python
import pandas as pd

arquivo = pd.read_excel("Downloads\cursinho.xlsx", header=1)
display(arquivo)
```

```python
import matplotlib.pyplot as plt

media_presentes = arquivo['Alunos presentes'].mean()

plt.figure(figsize=(8, 6))
plt.bar(arquivo['Matérias'], arquivo['Alunos presentes'], color='skyblue')
plt.axhline(media_presentes, color='red', linestyle='--', label=f'Média: {media_presentes:.2f}')
plt.title("Alunos Presentes por Matéria")
plt.xlabel("Matérias")
plt.ylabel("Número de Alunos Presentes")
plt.legend()
plt.show()
```

```python
media_faltas = arquivo['Alunos com falta'].mean()

plt.figure(figsize=(8, 6))
plt.bar(arquivo['Matérias'], arquivo['Alunos com falta'], color='lightcoral')
plt.axhline(media_faltas, color='blue', linestyle='--', label=f'Média: {media_faltas:.2f}')
plt.title("Alunos com Faltas por Matéria")
plt.xlabel("Matérias")
plt.ylabel("Número de Faltas")
plt.legend()
plt.show()
```

```python
alunos_presentes = arquivo['Alunos presentes']
dias_semana = arquivo['Dias da semana']

plt.title("Faltas por dia da semana")
plt.pie(alunos_presentes, labels = dias_semana, autopct='%1.2f%%')
plt.show()
```
